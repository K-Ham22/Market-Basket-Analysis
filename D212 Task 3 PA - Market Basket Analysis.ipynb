## **Data Transformation**

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
rows = []
for i in range(0, 7501):
    rows.append([str(df.values[i,j])
for j in range(0, 20)])
    
```

```python
#convert dataframe inot list of lists
rows = []
for i in range(0, 7501):
    rows.append([str(df.values[i,j])
    for j in range(0, 20)])
    
```

```python
#feed list into TransactionEncoder
TE = TransactionEncoder()
array = TE.fit(rows).transform(rows)

#return array to DataFrame
transaction = pd.DataFrame(array, columns = TE.columns_)
```

```python
transaction
```

```python
for col in transaction.columns:
    print(col)
```

```python
cleaned_df = transaction.drop(['nan'], axis=1)
cleaned_df.head()
```

```python
cleaned_df.shape
```

```python
for index, column_name in enumerate(cleaned_df.columns):
    print(f"Column name: {column_name}, Index: {index}")
```

```python
columns_to_drop = cleaned_df.columns[[0:556], [558:3615], [3616:4171], [4172:5283], [5284:7506]]
```

```python
indices_to_drop = [slice(0, 556), slice(558, 3615), slice(3616, 4171), slice(4172, 5283), slice(5284, 7506)]
```

```python
columns_to_drop = [c for i in indices_to_drop for c in cleaned_df.columns[i]]
df_dropped = cleaned_df.drop(columns=columns_to_drop)
```
